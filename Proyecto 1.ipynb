## Proyecto 1

```python
from bs4 import BeautifulSoup
import json

import requests
ciudades = ["maldonado", "montevideo"]
result = []
for ciudad in ciudades:
    
    url = f"https://braglia.com.uy/propiedades/listado/venta/apartamento/{ciudad}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    propiedades = soup.find_all('div', class_='estate-list-card')
    
    for propiedad in propiedades:
        descripcion = propiedad.find('h5', class_='estate-description')
        precio = propiedad.find("h2", class_= 'price-text-list')
        
        result.append({"tipo": descripcion.text.replace('\n', '').strip(), 
                       "precio": precio.text.replace('\n', '').strip(),
                       "ciudad":ciudad})

result
```

## Generación del archivo.json

```python
with open("propiedades.json", "w") as out_file:
    json.dump(result, out_file, indent=4)
```
